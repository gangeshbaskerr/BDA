```python
input= [
    "apple is a type of fruit.",
    "apple is in red color"
]

# Combine all input statements into a single string
combined_input = ' '.join(input)

# Split the combined input into words
words = combined_input.split()
def mapper(words):
    word_count = {}
    for word in words:
        word_count[word] = word_count.get(word, 0) + 1
    return word_count.items()
from collections import defaultdict

def reducer(mapped_items):
    word_count = defaultdict(int)
    for item in mapped_items:
        word, count = item
        word_count[word] += count
    return word_count.items()
# Step 1: Map
mapped_items = mapper(words)

# Step 2: Reduce
word_counts = reducer(mapped_items)

# Print the result
for word, count in word_counts:
    print(f'{word}: {count}')
```

```python
from collections import defaultdict

# Sample input with paragraphs labeled as p1, p2, etc.
input_text = {
    'p1': "This is the first paragraph. It contains some words.",
    'p2': "This is the second paragraph. It also contains some different words.",
    'p3': "Here is the third paragraph. It contains words that are both similar and different."
}

# Mapper function
def mapper(paragraphs):
    map_output = []
    
    for label, text in paragraphs.items():
        words = text.split()
        for word in words:
            # Emit (word, 1) for each word in each paragraph
            map_output.append((word, 1))
    
    return map_output

# Reducer function
def reducer(mapped_data):
    # Initialize a dictionary to hold the word counts
    reduced_data = defaultdict(int)
    
    # Aggregate counts for each word
    for word, count in mapped_data:
        reduced_data[word] += count
    
    return reduced_data

# Simulate the MapReduce process
def simulate_mapreduce(paragraphs):
    # Mapper phase
    mapped_data = mapper(paragraphs)
    
    # Shuffle and Sort phase (group by word)
    shuffled_data = defaultdict(list)
    for word, count in mapped_data:
        shuffled_data[word].append(count)
    
    # Prepare the data for the reducer
    reduced_input = [(word, sum(counts)) for word, counts in shuffled_data.items()]
    
    # Reducer phase
    reduced_output = reducer(reduced_input)
    
    return reduced_output

# Simulate MapReduce word count
word_count = simulate_mapreduce(input_text)

# Output the word count
for word, count in word_count.items():
    print(f"{word}: {count}")
```

```python
from collections import defaultdict

# Example matrices
A = [
    [1, 2],
    [3, 4],
    [5, 6]
]  # Matrix A (3x2)

B = [
    [7, 8, 9],
    [10, 11, 12]
]  # Matrix B (2x3)

# Matrix dimensions
m, n = len(A), len(A[0])
nB, p = len(B), len(B[0])

# Mapper function
def mapper(A, B):
    # Initialize map output
    map_output = []
    
    # Map for matrix A
    for i in range(m):
        for j in range(n):
            map_output.append((i, j, 'A', A[i][j]))
    
    # Map for matrix B
    for i in range(nB):
        for j in range(p):
            map_output.append((i, j, 'B', B[i][j]))
    
    return map_output

# Reducer function
def reducer(map_output):
    # Initialize intermediate storage
    intermediate = defaultdict(list)
    
    # Group by key (i, k) from map_output
    for i, j, mat, val in map_output:
        intermediate[(i, j)].append((mat, val))
    
    # Initialize the result matrix C with zeros
    C = [[0] * p for _ in range(m)]
    
    # Process the intermediate data
    for key in intermediate:
        i, j = key
        values = intermediate[key]
        
        if len(values) == 2:
            # Since values are mapped from two matrices A and B
            a_val = values[0][1] if values[0][0] == 'A' else values[1][1]
            b_val = values[0][1] if values[0][0] == 'B' else values[1][1]
            
            C[i][j] += a_val * b_val
    
    return C

# Execute MapReduce
map_output = mapper(A, B)
C = reducer(map_output)

# Print the resulting matrix C
print("Resulting Matrix C:")
for row in C:
    print(row)
```

```python
myfile = open("weather_data.txt")
txt = myfile.read()
months = []
min_temp = []
max_temp = []
total_list = txt.split(',')
total_list
for i in range(len(total_list)):
    if i % 3 == 0:
        months.append(total_list[i].strip("\n"))
    elif i % 3 == 1:
        min_temp.append(int(total_list[i]))
    else:
        max_temp.append(int(total_list[i]))
minimum = min(min_temp)
maximum = max(max_temp)
print("Month with minimum temperature of {t} is {m}".format(t=minimum, m=months[min_temp.index(minimum)]))
print("Month with maximum temperature of {t} is {m}".format(t =maximum, m= months[max_temp.index(maximum)]))
```
